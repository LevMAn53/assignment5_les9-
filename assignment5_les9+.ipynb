```python
import pandas as pd 
df = pd.read_excel("/Users/levmanulak/Desktop/ПОДКАСТ/data2.xls")
```

```python
df2 = df.iloc[:-1, :5]
```

```python
df_sorted = df2.iloc[:1001, :4]

# додав цю команду знизу, бо вибивало помилку, але не повністю розумію, що вона робить

pd.set_option('future.no_silent_downcasting', True)

df3 = df_sorted.fillna(df_sorted.mean()) 
```

```python
df_sorted2 = df2.iloc[1001:2001, :4] 

df_sorted2.ffill(inplace=True)
```

```python
df_sorted3 = df2.iloc[2001:, :4] 

df_sorted3.bfill(inplace=True)
```

```python
combined_df = pd.concat([df3, df_sorted2, df_sorted3])

#коли туту прописую combined_df.describe(), то мені показує лише дві колонки. 
#але коли просто роблю combined_df, то все окей. чому так?
```

```python
import sklearn

X = combined_df[['sepal_length', 'sepal_width', 'petal_length']]
Y = combined_df['petal_width']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)

# тут видає ось таку помилку: could not convert string to float: 'Random'
#не до кінця розумію, чому. 
```

```python
lr.predict(X_test)
pred = lr.predict(X_test)
pred
```
